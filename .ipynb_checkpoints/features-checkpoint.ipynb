```python
import numpy as np
import pandas as pd
from uszipcode import SearchEngine
search = SearchEngine(simple_zipcode = True)
import us_state_abbrev
statedict = us_state_abbrev.us_state_abbrev
```

```python
zori = pd.read_csv('./data/Zip_ZORI_AllHomesPlusMultifamily_SSA.csv')
zori = pd.concat([zori['RegionName'], zori.iloc[:, 4:]], axis = 1)
zori['CountyName'] = zori['RegionName'].apply(lambda zipcode: search.by_zipcode(zipcode).values()[5][:-7]\
                                              if search.by_zipcode(zipcode).values()[5] is not None else 'NA')
zori['StateName'] = zori['RegionName'].apply(lambda zipcode: search.by_zipcode(zipcode).values()[6]\
                                              if search.by_zipcode(zipcode).values()[6] is not None else 'NA')
zori['County'] = zori[['CountyName', 'StateName']].agg(', '.join, axis = 1)
zori = pd.concat([zori[['RegionName', 'County']], zori.iloc[:, 1:-3]], axis = 1)
zori.columns = zori.columns.str.replace('RegionName', 'ZipCode')
```

```python
def airqcompiler(start, end):
    final = pd.DataFrame()
    for i in range(start, end + 1):
        temp = pd.read_csv(f'./data/daily_aqi_by_county_{i}/daily_aqi_by_county_{i}.csv')[['county Name', 'State Name',
                                                                                    'Date', 'AQI']]
        temp['State Name'] = temp['State Name'].apply(lambda state: statedict.get(state, 'NA'))
        temp['County'] = temp[['county Name', 'State Name']].agg(', '.join, axis = 1)
        temp['Date'] = temp['Date'].apply(lambda date: pd.to_datetime(date, format = '%Y-%m-%d'))
        temp['Month'] = temp['Date'].apply(lambda date: date.month)
        temp['Year'] = temp['Date'].apply(lambda date: date.year)
        temp = temp.groupby(['County', 'Year', 'Month']).mean().reset_index()
        final = pd.concat([final, temp], axis = 0)
    return final
```

```python
airq = airqcompiler(2014, 2020)
```

```python
def popcompiler(start, end):
    final = pd.DataFrame()
    for i in range(start, end + 1):
        temp = pd.read_csv(f'./data/productDownload_2020-12-09T144241/ACSDT1Y{i}.B01003_data.csv', header = 1)
        temp = temp[['Geographic Area Name', 'Estimate!!Total']]
        temp.columns = ['County', 'Population']
        temp['County'] = temp['County'].apply(lambda county: county[:(county.find('County') - 1)] + ', ' +
                                              statedict.get(county[(county.find(',') + 2):], 'NA') if 'County'\
                                              in county else county[:(county.find('Municipio') - 1)] + ', ' +\
                                              statedict.get(county[(county.find(',') + 2):], 'NA'))
        temp['Year'] = [i for x in range(temp.shape[0])]
        final = pd.concat([final, temp], axis = 0)
    return final
```

```python
temp = popcompiler(2013, 2019)
population = pd.DataFrame()
for county, df in temp.groupby('County'): 
    df['Change'] = df['Population'].diff() / df['Population'].shift()
    df = df.iloc[np.arange(df.shape[0]).repeat(12)].reset_index().drop('index', axis = 1)
    df = pd.concat([df, pd.DataFrame(list(range(1, 13)) * int(df.shape[0] / 12), columns = ['Month'])], axis = 1)
    for year in pd.unique(df['Year'])[1:]:
        df.loc[df['Year'] == year, 'Population'] = np.array(df.loc[df['Year'] == (year - 1), 'Population']) * \
                                                   np.array((((1 + df.loc[df['Year'] == year, 'Change']) ** (1/12)) ** \
                                                               df.loc[df['Year'] == year, 'Month']))
    df = df[df['Year'] != 2013]
    df = df[['County', 'Year', 'Month', 'Population']]
    population = pd.concat([population, df], axis = 0)
```

```python
def CountyZORI(zipcode):
    temp = pd.DataFrame(zori[zori['ZipCode'] == zipcode].iloc[:, 2:].T.reset_index())
    temp.columns = ['Month', 'ZORI']
    temp['Year'] = temp['Month'].apply(lambda month: pd.to_datetime(month, format = '%Y-%m').year)
    temp['Month'] = temp['Month'].apply(lambda month: pd.to_datetime(month, format = '%Y-%m').month)
    temp['County'] = [zori[zori['ZipCode'] == zipcode]['County'].values[0] for _ in range(temp.shape[0])]
    return temp
```

```python
test = pd.merge(CountyZORI(10023), airq, on = ['County', 'Month', 'Year'], how = 'inner')
test = pd.merge(test, population, on = ['County', 'Month', 'Year'], how = 'inner')
test
```
